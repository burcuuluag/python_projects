```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("linear_regression_dataset.csv", sep=';')

plt.scatter(df.deneyim, df.maas)
plt.xlabel('deneyim')
plt.ylabel('maas')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

linear_reg = LinearRegression()

# verileri numpy array'ına çeviriyoruz. values yazmazsak tipi <class 'pandas.core.series.Series'> olacaktı
x = df.deneyim.values.reshape(-1, 1)
y = df.maas.values.reshape(-1, 1)

# modeli eğitiyoruz
linear_reg.fit(x, y)

# y eksenini kestiği nokta (intercept)
b0 = linear_reg.intercept_
print('b0: ', b0)

# eğim (slope)
b1 = linear_reg.coef_
print('b1: ', b1)

# yeni bir veri için tahmin yapma
maas_yeni = b0 + b1*11
print("yeni maas: ", maas_yeni)

# veya
print(linear_reg.predict([[11]]))

# regresyon doğrusunu çizdirme
plt.scatter(x, y)
plt.plot(x, linear_reg.predict(x), color='red')
plt.xlabel('deneyim')
plt.ylabel('maas')
plt.show()
```

# r^2 with linear regression:

```python
from sklearn.metrics import r2_score
```

```python
x = df.deneyim.values.reshape(-1, 1)
y = df.maas.values.reshape(-1, 1)

linear_reg.fit(x,y)

y_head = linear_reg.predict(x)

print("r2_score", r2_score(y,y_head))
```

r^2 değeri 0.97 dir. 1'e yakın çıktığı için lineer regresyon modelimizin performansı iyidir. lineer regresyon modeli iyi predict etmiştir.
