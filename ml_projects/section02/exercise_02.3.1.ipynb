```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

aşağıdaki kodda -5 ile 5 aralığında 51 tane sayı üretilir.

```python
x = np.linspace(-5, 5, 51)
x
```

(a)

x değerlerine karşılık gelir ve sigmoid fonksiyonunun formülü kullanılarak y değerleri hesaplanır.

```python
f_a = 1 / (1 + np.exp(-x))
f_a
```

x ve f_a değerleri kullanılarak grafik çizdirilir x ekseninin adı x, y ekseninin adı f(x)'dir

```python
plt.figure()
plt.plot(x, f_a)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

(b)

x değerlerine karşılık gelen f(x) değerleri hesaplanır. Bu grafikteki fonksiyon log-exp fonksiyonudur.

```python
f_b = np.log(1 + np.exp(x))
f_b
```

aşağıdaki kodda x ve f_b değerleriyle grafik çizdirilir.

```python
plt.figure()
plt.plot(x, f_b)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

(c)

2x2 alt grafiği olan bir obje oluşturulur ve 0. sıradaki subplot (yani soldaki) için f_a çizdirilirken, 1. sıradaki subplot (yani sağdaki) için f_b çizdirilir. Bu şekilde farklı grafikler bir arada gösterilir.

```python
# object oriented approach with 2x2 subplots
fig, ax = plt.subplots(1,2)
# add to ax object
ax[0].plot(x,f_a)
ax[1].plot(x,f_b)
plt.show()
```
