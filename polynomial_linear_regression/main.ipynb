```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("polynomial_regression.csv", sep=";")
df.head()
```

```python
from sklearn.preprocessing import PolynomialFeatures

df = pd.read_csv("polynomial_regression.csv", sep=";")

y = df.araba_max_hiz.values.reshape(-1,1)
x = df.araba_fiyat.values.reshape(-1,1)

plt.scatter(x,y)
plt.ylabel("araba_max_hiz")
plt.xlabel("araba_fiyat")

# PolynomialFeatures ile ikinci dereceden polinom fit ediyoruz.
poly_reg = PolynomialFeatures(degree=2)
x_poly = poly_reg.fit_transform(x)

# LinearRegression modelini kullanarak ikinci dereceden polinom fit ediyoruz.
lin_reg = LinearRegression()
lin_reg.fit(x_poly,y)

# Görselleştirme
y_head = lin_reg.predict(x_poly)
plt.plot(x, y_head, color="red")

plt.show()
```

```python
# assume engine size is 3.5L and horsepower is 350
print("10 milyon tl lik araba hizi tahmini: ",lin_reg.predict([[10000, 3.5, 350]]))
```

```python
# polynomial regression =  y = b0 + b1*x +b2*x^2 + b3*x^3 + ... + bn*x^n

from sklearn.preprocessing import PolynomialFeatures

#degree n değeri oluyor. yani modelimiz x^2 olacak.
polynomial_regression = PolynomialFeatures(degree = 4)

x_polynomial = polynomial_regression.fit_transform(x)
#print(x_polynomial)
linear_regression2 = LinearRegression()
linear_regression2.fit(x_polynomial,y)
```

```python
y_head2 = linear_regression2.predict(x_polynomial)

plt.plot(x,y_head2,color= "green",label = "poly")
plt.legend()
plt.show()
```
